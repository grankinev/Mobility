```python
def country_analysis(df_entity):
    #создаем сводную таблицу с пешеходами и водилами
    df_entity.drop(df_entity[df_entity['transportation_type'] == 'transit'].index, axis = 0, inplace = True)

    #переводим таблицу в удобный для работы вид
    dft = df_entity.T
    dft.drop(labels = ['geo_type', 'alternative_name'], axis = 0, inplace = True)
    dft.head()
    new_header = dft.iloc[0]
    dft = dft[1:]
    dft.columns = new_header

    dft.reset_index(inplace = True)
    dft.rename(columns={"index": "period"}, inplace = True)

    dft['period'].iloc[1:] = pd.to_datetime(dft['period'].iloc[1:], format = '%Y-%m-%d')


    #создаем сводную таблицу с пешеходами
    df_walkers = df_entity.drop(df_entity[(df_entity['transportation_type'] == 'driving') | (df_entity['transportation_type'] == 'transit')].index, axis = 0)


    #переводим таблицу в удобный для работы вид
    dft_walkers = df_walkers.T
    dft_walkers.drop(labels = ['geo_type','transportation_type', 'alternative_name'], axis = 0, inplace = True)

    new_header = dft_walkers.iloc[0]
    dft_walkers = dft_walkers[1:]
    dft_walkers.columns = new_header
    dft_walkers.reset_index(inplace = True)
    dft_walkers.rename(columns={"index": "period"}, inplace = True)
    dft_walkers['period'] = pd.to_datetime(dft_walkers['period'], format = '%Y-%m-%d')


    #создаем сводную таблицу с водилами
    df_drivers = df_entity.drop(df_entity[(df_entity['transportation_type'] == 'walking') | (df_entity['transportation_type'] == 'transit')].index, axis = 0)


    #переводим таблицу в удобный для работы вид
    dft_drivers = df_drivers.T
    dft_drivers.drop(labels = ['geo_type','transportation_type', 'alternative_name'], axis = 0, inplace = True)

    new_header = dft_drivers.iloc[0]
    dft_drivers = dft_drivers[1:]
    dft_drivers.columns = new_header
    dft_drivers.reset_index(inplace = True)
    dft_drivers.rename(columns={"index": "period"}, inplace = True)
    dft_drivers['period'] = pd.to_datetime(dft_drivers['period'], format = '%Y-%m-%d')

    # посчитаем среднее изменение активности на последней неделе с предпоследней
    activity_change = {}
    country_rising = []
    country_falling = []
    country_precrysis = {}
    country_deepcrysis = {}
    overall_activity_level = {}
    overall_activity_level_previous_7days = {}
    overall_activity_level_previous30 = {}
    country_deepcrysis_7daysago = {}
    country_precrysis_7daysago = {}
    number_of_improving = 0
    number_of_decelerating = 0

#     average_world_activity_lastdate = round(np.mean([dft_walkers.iloc[-7:,1:], dft_drivers.iloc[-7:,1:]]), 2)
#     average_world_activity_7daysago = round(np.mean([dft_walkers.iloc[-7][1:], dft_drivers.iloc[-7][1:]]), 2)
    pre_crysis_average = np.mean([np.mean(dft_walkers.iloc[1:60,:], axis = 0), np.mean(dft_drivers.iloc[1:60,:], axis = 0)])
    average_world_activity_lastdate = round(np.mean(([np.mean(dft_walkers.iloc[-7:,:], axis = 0), np.mean(dft_drivers.iloc[-7:,:], axis = 0)])), 2)
    average_world_activity_7daysago = round(np.mean(([np.mean(dft_walkers.iloc[-14:-7,:], axis = 0), np.mean(dft_drivers.iloc[-14:-7,:], axis = 0)])), 2)
    for column in dft_drivers.columns[1:]:
        # средняя за 7 дней по всем странам по водилам и пешеходам
        last_7days = np.mean([dft_drivers[column].iloc[-7:],dft_walkers[column].iloc[-7:]])
        previous_7days = np.mean([dft_drivers[column].iloc[-14:-7],dft_walkers[column].iloc[-14:-7]])
        overall_activity_level[column] = last_7days
        overall_activity_level_previous_7days[column] = previous_7days
        previous_30days = np.mean([dft_drivers[column].iloc[-37:-7],dft_walkers[column].iloc[-37:-7]])
        overall_activity_level_previous30[column] = previous_30days
        stdev = np.std(dft_drivers[column][:])

        if last_7days > 80:
            country_precrysis[column] = last_7days
        if last_7days < 0.3*pre_crysis_average:
            country_deepcrysis[column] = last_7days
        if previous_7days > 80:
            country_precrysis_7daysago[column] = previous_7days
        if previous_7days < 0.3*pre_crysis_average:
            country_deepcrysis_7daysago[column] = previous_7days           
            
        if last_7days > previous_30days:
            number_of_improving += 1
        else:
            number_of_decelerating += 1
     
            
    # Отрисуем график баров по миру
    overall_activity_level = OrderedDict(sorted(overall_activity_level.items(), key=itemgetter(1)))

    group_data = list(overall_activity_level.values())
    group_names = list(overall_activity_level.keys())
    group_data_30 = list(overall_activity_level_previous30.values())
    group_names_30 = list(overall_activity_level_previous30.keys())
    
    fig = plt.figure(figsize=(20,30))
    plt.style.use('fivethirtyeight')
    ax = fig.add_subplot(111)
    #fig, ax = plt.subplots()
    ax.barh(group_names, group_data, color = 'mediumblue', alpha = 0.75, edgecolor = None, label='Среднее значение за последнюю неделю')
    ax.barh(group_names_30, group_data_30, color = 'cornflowerblue',  alpha = 0.5, edgecolor = 'black',linewidth = 2, label='Среднее значение за предшествующие ей 30 дней')
    labels = ax.get_xticklabels()
    plt.setp(labels, rotation=0, horizontalalignment='right')
    ax.set(xlabel='Средний индекс активности', ylabel='Страна', title = 'Мобильность в 63 странах мира на ' + currentdate + '\n' + 'Среднемировой индекс мобильности за последние 7 дней: ' + str(average_world_activity_lastdate) 
           + ' (неделю назад: ' + str(average_world_activity_7daysago) + ')'
            + '\n' + 'Средний индекс мобильности в России за последние 7 дней: ' + str(round(overall_activity_level['Russia'],2)) + ' (неделю назад: ' + str(round(overall_activity_level_previous_7days['Russia'],2)) + ')'
           + '\n' + 'Число стран, в которых активность за неделю выше, чем за месяц до нее: ' + str(number_of_improving)
           + '\n' + 'Число стран, в которых активность за неделю уменьшилась по сравнению с месяцем назад: ' + str(number_of_decelerating)
          + '\n' + 'Число стран, переживающих самый глубокий карантин (мобильность меньше 30% докризисной): ' + str(len(country_deepcrysis)) + ' из 63')


    labels = ax.get_xticklabels()
    ax.axvline(pre_crysis_average, ls='--', color='green', label='Докризисный уровень мобильности в мире \n (среднее за 13.01 - 13.03)')
    ax.axvline(0.5*pre_crysis_average, ls='--', color='orange', label='50% от докризисного уровня')
    ax.axvline(0.3*pre_crysis_average, ls='--', color='red', label='30% от докризисного уровня')

    plt.legend(loc=5)
    pic_name = 'C:\\Users\\Egran\\YandexDisk\\Work\\Различные тесты\\Mobility\\charts\\' + 'countries_overview_' + str(currentdate) + '.jpg'
    plt.savefig(pic_name,bbox_inches='tight', pad_inches=0)



```

```python
def city_analysis(df_entity):
    #создаем сводную таблицу с пешеходами и водилами
    df_entity.drop(df_entity[df_entity['transportation_type'] == 'transit'].index, axis = 0, inplace = True)

    #переводим таблицу в удобный для работы вид
    dft = df_entity.T
    dft.drop(labels = ['geo_type', 'alternative_name'], axis = 0, inplace = True)
    dft.head()
    new_header = dft.iloc[0]
    dft = dft[1:]
    dft.columns = new_header

    dft.reset_index(inplace = True)
    dft.rename(columns={"index": "period"}, inplace = True)

    dft['period'].iloc[1:] = pd.to_datetime(dft['period'].iloc[1:], format = '%Y-%m-%d')


    #создаем сводную таблицу с пешеходами
    df_walkers = df_entity.drop(df_entity[(df_entity['transportation_type'] == 'driving') | (df_entity['transportation_type'] == 'transit')].index, axis = 0)


    #переводим таблицу в удобный для работы вид
    dft_walkers = df_walkers.T
    dft_walkers.drop(labels = ['geo_type','transportation_type', 'alternative_name'], axis = 0, inplace = True)

    new_header = dft_walkers.iloc[0]
    dft_walkers = dft_walkers[1:]
    dft_walkers.columns = new_header
    dft_walkers.reset_index(inplace = True)
    dft_walkers.rename(columns={"index": "period"}, inplace = True)
    dft_walkers['period'] = pd.to_datetime(dft_walkers['period'], format = '%Y-%m-%d')


    #создаем сводную таблицу с водилами
    df_drivers = df_entity.drop(df_entity[(df_entity['transportation_type'] == 'walking') | (df_entity['transportation_type'] == 'transit')].index, axis = 0)


    #переводим таблицу в удобный для работы вид
    dft_drivers = df_drivers.T
    dft_drivers.drop(labels = ['geo_type','transportation_type', 'alternative_name'], axis = 0, inplace = True)

    new_header = dft_drivers.iloc[0]
    dft_drivers = dft_drivers[1:]
    dft_drivers.columns = new_header
    dft_drivers.reset_index(inplace = True)
    dft_drivers.rename(columns={"index": "period"}, inplace = True)
    dft_drivers['period'] = pd.to_datetime(dft_drivers['period'], format = '%Y-%m-%d')

    # посчитаем среднее изменение активности на последней неделе с предпоследней
    activity_change = {}
    city_rising = []
    city_falling = []
    city_precrysis = {}
    city_deepcrysis = {}
    overall_activity_level = {}
    overall_activity_level_previous30 = {}
    number_of_improving = 0
    number_of_decelerating = 0

    #average_world_activity_lastdate = round(np.mean([dft_walkers.iloc[-1][1:], dft_drivers.iloc[-1][1:]]), 2)
    average_world_activity_lastdate = np.mean([np.mean(dft_walkers.iloc[-7:,:], axis = 1), np.mean(dft_drivers.iloc[-7:,:], axis = 1)])
    average_world_activity_lastdate = round(average_world_activity_lastdate,2)
    pre_crysis_average = np.mean([np.mean(dft_walkers.iloc[1:60,:], axis = 1), np.mean(dft_drivers.iloc[1:60,:], axis = 1)])

    for column in dft_drivers.columns[1:]:
        # средняя за 7 дней по всем странам по водилам и пешеходам
        last_7days = np.mean([dft_drivers[column].iloc[-7:],dft_walkers[column].iloc[-7:]])
        overall_activity_level[column] = last_7days
        previous_30days = np.mean([dft_drivers[column].iloc[-37:-7],dft_walkers[column].iloc[-37:-7]])
        overall_activity_level_previous30[column] = previous_30days
        stdev = np.std(dft_drivers[column][:])
        if dft_drivers[column].iloc[-1] > last_7days:
            city_rising.append(column)
        if dft_drivers[column].iloc[-1] < last_7days:
            city_falling.append(column)

        if last_7days > 80:
            city_precrysis[column] = last_7days
        if last_7days < 0.3*pre_crysis_average:
            city_deepcrysis[column] = last_7days
            
        if last_7days > previous_30days:
            number_of_improving += 1
        else:
            number_of_decelerating += 1
            
            
    # Отрисуем график баров по миру
    overall_activity_level = OrderedDict(sorted(overall_activity_level.items(), key=itemgetter(1)))

    group_data = list(overall_activity_level.values())
    group_names = list(overall_activity_level.keys())
    group_data_30 = list(overall_activity_level_previous30.values())
    group_names_30 = list(overall_activity_level_previous30.keys())

    fig = plt.figure(figsize=(20,35))
    plt.style.use('fivethirtyeight')
    ax = fig.add_subplot(111)
    #fig, ax = plt.subplots()
    ax.barh(group_names, group_data, color = 'mediumblue', alpha = 0.75, edgecolor = None, label='Среднее значение за последнюю неделю')
    ax.barh(group_names_30, group_data_30, color = 'cornflowerblue',  alpha = 0.5, edgecolor = 'black',linewidth = 2, label='Среднее значение за предшествующие ей 30 дней')
    labels = ax.get_xticklabels()
    plt.setp(labels, rotation=0, horizontalalignment='right')
    ax.set(xlabel='Средний индекс активности', ylabel='Город', title = 'Мобильность в 89 городах мира на ' + currentdate + '\n' + 'Среднемировой индекс мобильности по городам: ' + str(average_world_activity_lastdate)
            + '\n' + 'Средний индекс мобильности в Москве за последние 7 дней: ' + str(round(overall_activity_level['Moscow'],2))
           + '\n' + 'Число городов, в которых активность за неделю выше, чем за месяц до нее: ' + str(number_of_improving)
           + '\n' + 'Число городов, в которых активность за неделю уменьшилась по сравнению с месяцем назад: ' + str(number_of_decelerating)
          + '\n' + 'Число городов, переживающих самый глубокий карантин (мобильность меньше 30% докризисной): ' + str(len(city_deepcrysis)) + ' из ' + str(len(dft_drivers.columns[1:]))
           
          )
    ax.axvline(pre_crysis_average, ls='--', color='green', label='Докризисный уровень мобильности в среднем по городам \n (среднее за 13.01 - 13.03)')
    ax.axvline(0.5*pre_crysis_average, ls='--', color='orange', label='50% от докризисного уровня')
    ax.axvline(0.3*pre_crysis_average, ls='--', color='red', label='30% от докризисного уровня')

    plt.legend(loc=5)
    pic_name = 'C:\\Users\\Egran\\YandexDisk\\Work\\Различные тесты\\Mobility\\charts\\' + 'cities_overview_' + str(currentdate) + '.jpg'
    plt.savefig(pic_name,bbox_inches='tight', pad_inches=0)



```

```python
df_entity = df_countries
#создаем сводную таблицу с пешеходами и водилами
df_entity.drop(df_entity[df_entity['transportation_type'] == 'transit'].index, axis = 0, inplace = True)

#переводим таблицу в удобный для работы вид
dft = df_entity.T
dft.drop(labels = 'geo_type', axis = 0, inplace = True)
dft.head()
new_header = dft.iloc[0]
dft = dft[1:]
dft.columns = new_header

dft.reset_index(inplace = True)
dft.rename(columns={"index": "period"}, inplace = True)

dft['period'].iloc[1:] = pd.to_datetime(dft['period'].iloc[1:], format = '%Y-%m-%d')


#создаем сводную таблицу с пешеходами
df_walkers = df_entity.drop(df_entity[(df_entity['transportation_type'] == 'driving') | (df_entity['transportation_type'] == 'transit')].index, axis = 0)


#переводим таблицу в удобный для работы вид
dft_walkers = df_walkers.T
dft_walkers.drop(labels = ['geo_type','transportation_type'], axis = 0, inplace = True)

new_header = dft_walkers.iloc[0]
dft_walkers = dft_walkers[1:]
dft_walkers.columns = new_header
dft_walkers.reset_index(inplace = True)
dft_walkers.rename(columns={"index": "period"}, inplace = True)
dft_walkers['period'] = pd.to_datetime(dft_walkers['period'], format = '%Y-%m-%d')


#создаем сводную таблицу с водилами
df_drivers = df_entity.drop(df_entity[(df_entity['transportation_type'] == 'walking') | (df_entity['transportation_type'] == 'transit')].index, axis = 0)


#переводим таблицу в удобный для работы вид
dft_drivers = df_drivers.T
dft_drivers.drop(labels = ['geo_type','transportation_type'], axis = 0, inplace = True)

new_header = dft_drivers.iloc[0]
dft_drivers = dft_drivers[1:]
dft_drivers.columns = new_header
dft_drivers.reset_index(inplace = True)
dft_drivers.rename(columns={"index": "period"}, inplace = True)
dft_drivers['period'] = pd.to_datetime(dft_drivers['period'], format = '%Y-%m-%d')

# посчитаем среднее изменение активности на последней неделе с предпоследней
activity_change = {}
country_rising = []
country_falling = []
country_precrysis = {}
country_deepcrysis = {}
overall_activity_level = {}
overall_activity_level_previous30 = {}

average_world_activity_lastdate = round(np.mean([dft_walkers.iloc[-1][1:], dft_drivers.iloc[-1][1:]]), 2)
pre_crysis_average = np.mean([np.mean(dft_walkers.iloc[1:60,:], axis = 0), np.mean(dft_drivers.iloc[1:60,:], axis = 0)])

for column in dft_drivers.columns[1:]:
    # средняя за 7 дней по всем странам по водилам и пешеходам
    last_7days = np.mean(dft_drivers[column].iloc[-7:])
    overall_activity_level[column] = last_7days
    previous_30days = np.mean(dft_drivers[column].iloc[-37:-7])
    overall_activity_level_previous30[column] = previous_30days
    stdev = np.std(dft_drivers[column][:])
    if dft_drivers[column].iloc[-1] > last_7days:
        country_rising.append(column)
    if dft_drivers[column].iloc[-1] < last_7days:
        country_falling.append(column)

    if last_7days > 80:
        country_precrysis[column] = last_7days
    if last_7days < 0.3*pre_crysis_average:
        country_deepcrysis[column] = last_7days

# Отрисуем график баров по миру
overall_activity_level = OrderedDict(sorted(overall_activity_level.items(), key=itemgetter(1)))

group_data = list(overall_activity_level.values())
group_names = list(overall_activity_level.keys())
group_data_30 = list(overall_activity_level_previous30.values())
group_names_30 = list(overall_activity_level_previous30.keys())

fig = plt.figure(figsize=(20,30))
plt.style.use('fivethirtyeight')
ax = fig.add_subplot(111)
#fig, ax = plt.subplots()
ax.barh(group_names, group_data, color = 'mediumblue', alpha = 0.75, label='Среднее значение за последнюю неделю')
ax.barh(group_names_30, group_data_30, color = 'b', alpha = 0.3, label='Среднее значение за предшествующие ей 30 дней')
labels = ax.get_xticklabels()
plt.setp(labels, rotation=0, horizontalalignment='right')
ax.set(xlabel='Средний индекс активности', ylabel='Страна',
       title='Средняя мобильность в мире по водителям на ' + currentdate + '\n' + 'Индекс мировой мобильности: ' + str(average_world_activity_lastdate)
      + '\n' + 'Число стран, переживающих самый глубокий карантин (мобильность меньше 30% докризисной): ' + str(len(country_deepcrysis)) + ' из 63'
      + '\n' + 'Уровень мобильности в России за последние 7 дней: ' + str(round(overall_activity_level['Russia'],2)))

ax.axvline(pre_crysis_average, ls='--', color='green', label='Докризисный уровень мобильности в мире \n (среднее за 13.01 - 13.03)')
ax.axvline(0.5*pre_crysis_average, ls='--', color='orange', label='50% от докризисного уровня')
ax.axvline(0.3*pre_crysis_average, ls='--', color='red', label='30% от докризисного уровня')

ax.text(0.95, 0.01, 'Источник: Apple, оценки ЦЭП ГПБ',
        verticalalignment='bottom', 
        transform=ax.transAxes,
        color='black', fontsize=8)



plt.legend(loc=5)
fig_name = 'drivers_overview_' + str(currentdate) + '.jpg'
plt.savefig(fig_name,bbox_inches='tight', pad_inches=0)



```
